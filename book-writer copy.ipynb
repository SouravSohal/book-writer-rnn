```python
import tensorflow as tf
import numpy as np
import os
```

```python
text= open(file, 'rb').read().decode(encoding='utf-8')
```

```python
vocab = sorted(set(text.split()))
```

```python
len(vocab)
```

```python
w2i= {a:i for i, a in enumerate(vocab)}
```

```python
i2w=np.array(vocab)
```

```python
text_as_seq_of_ints= np.array([w2i[a] for a in text.split()])
```

```python
text_as_seq_of_ints
```

```python
seq_length = 30
examples_per_epoch = len(text.split())//(seq_length+1)
```

```python
word_dataset = tf.data.Dataset.from_tensor_slices(text_as_seq_of_ints)
```

```python
for i in word_dataset.take(20):
    print(i2w[i.numpy()], i.numpy())
```

```python
seq=word_dataset.batch(seq_length+1, drop_remainder=True)
```

```python
for i in seq.take(2):
    print(i2w[i.numpy()])
```

```python
def f_make_input_target_pairs(s):
    input_text = s[:-1]
    target_text = s[1:]
    return input_text, target_text
```

```python
dataset= seq.map(f_make_input_target_pairs)
```

```python
for X,y in dataset.take(1):
    print('Input:', ' '.join(i2w[X.numpy()]))
    print('Target:', ' '.join(i2w[y.numpy()]))
```

```python
BS= 50
dataset=dataset.shuffle(1000).batch(BS,drop_remainder=True)
```

```python
VS= len(vocab)
ED= 100
NU= 200
```

```python
def f_make_model(VS, ED, NU, BS):
    inputs = tf.keras.Input(batch_shape=(BS, None), dtype=tf.int32)
    x = tf.keras.layers.Embedding(VS, ED)(inputs)
    x = tf.keras.layers.GRU(NU, return_sequences=True, stateful=True)(x)
    outputs = tf.keras.layers.Dense(VS)(x)
    return tf.keras.Model(inputs, outputs)
```

```python
model= f_make_model(VS, ED, NU, BS)
```

```python
def f_loss(y, y_hat):
    return tf.keras.losses.sparse_categorical_crossentropy(y,y_hat,from_logits=True)
```

```python
model.compile(optimizer='adam', loss=f_loss)
```

```python
checkpoints_dir= './tr_checkpoints'
checkpoint_prefix=os.path.join(checkpoints_dir,"chpt_{epoch}.weights.h5")
checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix, 
    save_weights_only=True
)
```

```python
for x, y in dataset.take(2):
    print("Input shape:", x.shape)
    print("Target shape:", y.shape)
```

```python
history=model.fit(dataset,epochs=15, callbacks=[checkpoint_callback])
```
